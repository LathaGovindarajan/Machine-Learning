```r
install.packages("igraph")
library(igraph)
```

```r
nodes=data.frame(name=c("latha","ashu","lax","vichu"),gender=c("f","f","f","m"),age=c(22,22,21,22))
```

```r
nodes
```

```r
edges=data.frame(from=c("latha","lax","vichu","ashu","latha"),to=c("ashu","ashu","latha","vichu","lax"))
```

```r
edges
```

```r
g<-graph.data.frame(edges,directed = FALSE,vertices = nodes)
plot(g)
```

```r
layout1<-g%>% layout_nicely()
g%>%plot(vertext_size=50,layout=layout1)
```

```r
colors<-ifelse(V(g)$gender=="m","lavender","pink")
g%>%plot(vertext_size=30,vertex.color=colors,layout=layout1)
```

```r
V(g)
E(g)
```

```r
g
```

```r
friends<-ego(g,order=1,nodes="latha",mindist=1)
friends
```

```r
degree<-g%>%degree()%>%print()
```

```r
closeness<-g%>%closeness()%>%print()
```

```r
betweenness<-g%>%betweenness()%>%print()
```
